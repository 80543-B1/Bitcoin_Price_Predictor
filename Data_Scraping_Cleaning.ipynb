# Web Scraping

## Scraping the following CryptoCurrency data from investing.com  (https://in.investing.com/crypto/)

* Bitcoin
* BNB
* DogeCoin
* Ethereum
* Tether
* LiteCoin
* USDCoin

```python
from bs4 import BeautifulSoup

with open("bitcoin.html","r+") as file:
    data_btc = file.read()
with open("BNB.html","r+") as file:
    data_bnb = file.read()
with open("Dogecoin.html","r+") as file:
    data_doge = file.read()
with open("Ethereum.html","r+") as file:
    data_eth = file.read()
with open("Tether.html","r+") as file:
    data_teth = file.read()
with open("LiteCoin.html","r+") as file:
    data_lite = file.read()
with open("USD Coin.html","r+") as file:
    data_usdc = file.read()

Files = [data_btc , data_bnb , data_doge , data_eth , data_teth , data_lite , data_usdc ]
```

```python
#  function to write the values in the csv file

def write_data(row,file):
    columns = row.find_all("td")
    file.write("\n")
    for i in range(len(columns)):
        file.write(str(columns[i].text))
        file.write(",")
    return file
```

```python
for file in Files:
    soup = BeautifulSoup(file,"html.parser")
    tbody = soup.find("tbody")
    rows = tbody.find_all("tr")
    if file ==  data_btc:
        for row in rows:
            file = open("Uncleaned_Bitcoin.csv", "a+")
            write_data(row,file)
    if file ==  data_bnb:
        for row in rows:
            file = open("Uncleaned_BNB.csv", "a+")
            write_data(row,file)

    if file ==  data_doge:
        for row in rows:
            file = open("Uncleaned_Dogecoin.csv", "a+")
            write_data(row,file)

    if file ==  data_eth:
        for row in rows:
            file = open("Uncleaned_Ethereum.csv", "a+")
            write_data(row,file)

    if file == data_teth:
        for row in rows:
            file = open("Uncleaned_Tether.csv", "a+")
            write_data(row,file)

    if file ==  data_lite:
        for row in rows:
            file = open("Uncleaned_Litecoin.csv", "a+")
            write_data(row,file)

    if file ==  data_usdc:
        for row in rows:
            file = open("Uncleaned_USDCoin.csv", "a+")
            write_data(row,file)
    
```

# Data Cleaning

## Importing the edited csv files

```python
df = pd.read_csv("Cleaned_Bitcoin.csv")
df
```

```python
df.info()
```

```python
df.rename(columns = { 
    "Vol." : "Volume",
    "Change %" : "Change"
},inplace = True)
```

```python
df
```

```python
# removing the commas from price , Open , High , Low

df["Price"] = df["Price"].str.replace(',', '').astype(float)
df["Open"] = df["Open"].str.replace(',', '').astype(float)
df["High"] = df["High"].str.replace(',', '').astype(float)
df["Low"] = df["Low"].str.replace(',', '').astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
# converting the date object into DateTime Format since we dealing with timeseries data

df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
# removing % icon from every record in df["Change"]

df["Change"] = df["Change"].str.replace("%" , "").astype("float")
```

```python
df.head(3)
```

```python
# Converting the "K" , "M" and "B" string values into numerical in df["Volume"] 

df.dropna(inplace = True)
```

```python
df["Volume"] = df["Volume"].replace({ "K" : "*1e3" , "M" : "*1e6" , "B" : "*1e9" } , regex = True ).map(pd.eval).astype(float)
```
